Step 1 â€“ Data Loading & Import necessary libraries




```python
#Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('Mall_Customers.csv')
```

Cluster customers into segments based on income and spending score

```python
# 1. Select features for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# 2. Use the Elbow Method to find the optimal number of clusters
inertia = []
K = range(1, 11)

for k in K:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X)
    inertia.append(model.inertia_)

# 3. Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method to Determine Optimal k')
plt.grid(True)
plt.show()



```

Perform scaling and visual exploration of groupings- Apply K-Means clustering and determine optimal number of clusters

```python
# 1. Select relevant features for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# 2. Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Apply KMeans with the chosen number of clusters (e.g., k = 4)
k = 4
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# 4. Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='Set2', s=100)
plt.title('Customer Segments based on Income & Spending')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.legend(title='Cluster')
plt.show()
```

Visualize clusters using 2D plots

```python
# 1. Plot the clusters
plt.figure(figsize=(8, 6))

# Use the original (unscaled) values for easier interpretation
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'],
            c=df['Cluster'], cmap='viridis', s=50, alpha=0.7)

# 2. Label axes and title
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segments (K-Means Clustering)')
plt.grid(True)
plt.colorbar(label='Cluster')
plt.show()
```

Try DBSCAN clustering algorithm

```python
from sklearn.cluster import DBSCAN


# 1. Select features
X_dbscan = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# 2. Standardize the data (important for DBSCAN)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_dbscan)

# 3. Apply DBSCAN
db = DBSCAN(eps=0.5, min_samples=5)
db_labels = db.fit_predict(X_scaled)

# 4. Add DBSCAN cluster labels to the DataFrame
df['DBSCAN_Cluster'] = db_labels

# 5. Visualize DBSCAN clusters
plt.figure(figsize=(8, 6))
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'],
            c=db_labels, cmap='tab10', s=50, alpha=0.7)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('DBSCAN Clustering of Customers')
plt.colorbar(label='Cluster Label')
plt.grid(True)
plt.show()
```

Analyze average spending per cluster

```python
# 1. K-Means: Average Spending Score per Cluster
kmeans_avg = df.groupby('Cluster')['Spending Score (1-100)'].mean().reset_index()
print("Average Spending Score per K-Means Cluster:")
print(kmeans_avg)

# 2. DBSCAN: Average Spending Score per Cluster (excluding outliers if needed)
dbscan_avg = df[df['DBSCAN_Cluster'] != -1].groupby('DBSCAN_Cluster')['Spending Score (1-100)'].mean().reset_index()
print("\n Average Spending Score per DBSCAN Cluster (excluding noise):")
print(dbscan_avg)
```
